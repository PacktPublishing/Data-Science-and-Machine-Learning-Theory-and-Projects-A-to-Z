```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import cv2
```

```python
im = plt.imread(r'c:\Users\Dell\Pictures\Saved Pictures\bird.png')
```

```python
im.shape
```

```python
im2 = im[:,:,:3]
```

```python
plt.imshow(im2)
```

```python
imGray = cv2.cvtColor(im2,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(imGray,cmap='gray')
```

```python
mask = np.ones((19,19))/(19**2)
```

```python
imBlurred = signal.convolve2d(imGray,mask,boundary='symm',mode='same')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10,15))
ax1.imshow(imGray,cmap='gray')
ax2.imshow(imBlurred,cmap='gray')
fig.show()
```
