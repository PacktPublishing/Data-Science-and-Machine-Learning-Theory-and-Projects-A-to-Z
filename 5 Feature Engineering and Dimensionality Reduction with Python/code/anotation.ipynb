```python
import plotly.graph_objs as go
import numpy as np
X,Y = np.meshgrid(np.linspace(-2,2,30),np.linspace(-2,2,30))
Z = np.cos(X**2+Y**2)
t = 1
trace = go.Scatter3d(x=X.flatten()[::t],
                    y = Y.flatten()[::t],
                    z = Z.flatten()[::t],mode='markers')
fig = go.Figure(data = [trace])
fig.show()
```
