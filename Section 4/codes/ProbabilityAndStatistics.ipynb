```python
def BernoulliTrail(p=0.5):
    X = int(np.random.rand()<=p)
    return X
```

```python
print(BernoulliTrail(0.7))
```

```python
countOnes = 0
countZeros = 0
p = 0.611334
n = 1000000
for i in range(n):
    x = BernoulliTrail(p)
    if x == 1:
        countOnes += 1
    else:
        countZeros += 1
print("Success Fraction is : ",countOnes/n)
```

```python
def GT(p):
    X = 1
    while True:
        if BernoulliTrail(p) == 1:
            return X
        else:
            X+=1
```

```python
p = 0.2
print(GT(p))
```

```python
n = 1000
p = 0.6
G = np.zeros(n)
for i in range(n):
    G[i] = GT(p)
```

```python
# E[X] = p
p = 0.812
sMeanX = []
for N in range(1,10000,100):
    sMeanX.append((np.random.rand(N)<=p).mean())
y = np.array(sMeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
```

# Geometric(p), $E[X]=1/p$

```python
# E[X] = 1/p

p = 0.312
print(1/p)
sMeanX = []
for N in range(1,10000,100):
    sMeanX.append((np.random.geometric(p,N).mean()))
y = np.array(sMeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
```
